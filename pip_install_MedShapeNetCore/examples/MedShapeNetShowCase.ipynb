```python
from MedShapeNetCore.MedShapeNetCore import MyDict,MSNLoader,MSNVisualizer,MSNSaver,MSNTransformer
msn_loader=MSNLoader()
msn_vis=MSNVisualizer()
```

```python
ThoracicAorta_Saitta_data=msn_loader.load('ThoracicAorta_Saitta')
```

```python
AVT_data=msn_loader.load('AVT')
```

```python
points_AVT=AVT_data['point']
print(len(points_AVT))
```

```python
CoronaryArteries_data=msn_loader.load('CoronaryArteries')
```

```python
points_ThoracicAorta=ThoracicAorta_Saitta_data['point']
print(len(points_ThoracicAorta))

points_CoronaryArteries=CoronaryArteries_data['point']
print(len(points_CoronaryArteries))
```

```python
%matplotlib notebook
```

```python
msn_vis.plot_point(points_ThoracicAorta[0][:,0],points_ThoracicAorta[0][:,1],points_ThoracicAorta[0][:,2])
```

```python
msn_vis.plot_point(points_CoronaryArteries[0][:,0],points_CoronaryArteries[0][:,1],points_CoronaryArteries[0][:,2])
```

```python
msn_vis.plot_point(points_AVT[0][:,0],points_AVT[0][:,1],points_AVT[0][:,2])
```

```python
TeethSeg_data=msn_loader.load('3DTeethSeg') # from 3DTeethSeg challenge
```

```python
TeethSeg_data['0132CR0A'].keys()  
# the avilable keys (like '0132CR0A') correspond to the
# patient ID of the original 3DTeethSeg challenge dataset
```

```python
TeethSeg_data['0132CR0A']['mesh'].keys() 
```

```python
print(TeethSeg_data['0132CR0A']['mesh']['lower'].keys())
print(TeethSeg_data['0132CR0A']['mesh']['upper'].keys())
```

```python
v_lower=TeethSeg_data['0132CR0A']['mesh']['lower']['vertices']
f_lower=TeethSeg_data['0132CR0A']['mesh']['lower']['faces']
v_upper=TeethSeg_data['0132CR0A']['mesh']['upper']['vertices']
f_upper=TeethSeg_data['0132CR0A']['mesh']['upper']['faces']

print(v_lower.shape)
print(f_lower.shape)
print(v_upper.shape)
print(f_upper.shape)
```

```python

msn_vis.plot_point(v_lower[:,0],v_lower[:,1],v_lower[:,2])
```
